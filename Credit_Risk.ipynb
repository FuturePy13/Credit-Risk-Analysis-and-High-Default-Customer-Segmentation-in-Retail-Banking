# Credit Risk Analysis and High-Default Customer Segmentation in Retail Banking

## Business Context

* Loan delinquency refers to a delay in the repayment of dept acording to agreed terms. 
* Over the past 4 months, the bank has experienced a consistent increase in delinquency rates whitin its personal loan portfolio. This trend is negatively impacting profiability due to reduced cash inflows and higher credit risk exposure. 
* If not adressed, this could force the bank to tighten lending conditions or increase interest rates to compensate for losses, potentially reducing loan demand and limiting future growth. 

### DataFrame 

```python
import pandas as pd
import numpy as np

np.random.seed(42)
# Customer Creation
n = 8000
data = pd.DataFrame({
    "customer_id" : range(1, n + 1),
    "age" : np.random.randint(21, 65, n),
    "monthly_income" : np.random.normal(2500, 800, n).clip(500, 10000),
    "credit_score" : np.random.normal(650, 70, n).clip(300, 850)
})

# Credit
data['loan_amount'] = np.random.normal(5000, 2000, n).clip(500, 20000)
data['loan_term_months'] = np.random.choice([12, 24, 36, 48], n)
data['interest_rate'] = np.random.uniform(0.1, 0.3, n)

# Risk Logic
risk_score = (
    (data['loan_amount'] / data['monthly_income']) * 0.5 +
    (700 - data['credit_score']) * 0.003
)

prob_default = 1 / (1 + np.exp(-risk_score))

data['default'] = np.random.binomial(1, prob_default)

# Categorize Status
data['loan_status'] = data['default'].map({
    0: "Current",
    1: "Default"
})
```
